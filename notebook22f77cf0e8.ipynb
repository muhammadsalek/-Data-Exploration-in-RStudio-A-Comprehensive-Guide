```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 20)
y = np.sin(x)
y2 = np.cos(x)

# Line plot with two lines
plt.plot(x, y, label='sin(x)', marker='o')   # sine line
plt.plot(x, y2, label='cos(x)', marker='s')  # cosine line

plt.title("Sine and Cosine Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()   # legend  
plt.grid(True) # grid lines  
plt.show()
```

```python
# 2. Line plot multiple lines
plt.plot(x, y, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```

```python
# 3. Scatter plot simple
plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.show()
```

```python
# 5. Scatter plot sized
sizes = np.random.randint(20, 200, size=len(x))
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title("Sized Scatter Plot")
plt.show()
```
